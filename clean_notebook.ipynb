```python
import json

def find_clean(content):
    for key in content.keys():
        if key == "text":
            content[key] = ""
        if key == "outputs":
            content[key] = []
        if type(content[key]) == dict:
            find_clean(content[key])
```

```python
cleaned_content = json.loads(open("./BaseModelGenerator.ipynb", "r").read())
content_cells = cleaned_content['cells']

for i in range(len(content_cells)):
    find_clean(content_cells[i])

cleaned_content['cells'] = content_cells

with open("./BaseModelGeneratorCleaned.ipynb", "w") as f:
    f.write(json.dumps(cleaned_content))
```
